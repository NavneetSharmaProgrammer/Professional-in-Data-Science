# Loops in Python
#### Python has two types of loops : <b>For<b> and <b>while<b>.

# 1. For Loop 
#### Used to iterate over sequence like list, tuples, and strings.

```python
fruits = {"apple","banana","cherry"}
```

```python
    for fruit in fruits:
        print(fruit)
```

## Using range()

```python
    for i in range(3): # default (start, stop,step) default (0,3,1)
        print(i)
```

```python
    for i in range(1,4,2):
        print(i)
```

# 2. While Loop
#### Runs as long as condition is <b>True<b>.

```python
count = 0
while count<3:
    print(count)
    count += 1
```

```python
i = 0
while i<12:
    print(i)
    i += 1
```

# 3. Loop Control Statements
#### break → Exits the loop.
#### continue → Skips to the next iteration.
#### pass → Does nothing (used as a placeholder).

```python
l = {"apple","banana","cherry"}
for item in l:
    pass # Dont do anything
    
print("Hello")
```

```python
for item in l:
    if item == "banana":
        break # Stop this loop right here
    print(item)
```

```python
for item in l:
    if item == "banana":
        continue # Skip this perticular iteration of the loop
    print(item)
```
